```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('training.txt')
data
```

```python
def suffle(x):
    arr = np.arange(0, len(x))
    np.random.shuffle(arr)
    return x[arr]   
```

```python
suffled_data = suffle(data)
suffled_data
```

```python
splited_data = np.split(suffled_data, (2, 3), axis=1)
X = splited_data[0]
y = splited_data[1]
X
```

```python
from layer import BasicNeuralNetwork

in_dim_n = 2 
hidden_dim_n = 8 
out_dim_n = 1
epoch = 10
learning_rate = 0.01
batch_size = 1

bnn = BasicNeuralNetwork(
    in_dim_n, hidden_dim_n, out_dim_n, epoch, learning_rate, batch_size)
```

```python
result = bnn.train(X, y)
plt.plot(result)
```
