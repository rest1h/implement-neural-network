```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('training.txt')
data
```

```python
val_data = np.loadtxt('test.txt')
val_data
```

```python
def suffle(x):
    arr = np.arange(0, len(x))
    np.random.shuffle(arr)
    return x[arr]   
```

```python
suffled_data = suffle(data)
suffled_data
```

```python
suffled_val_data = suffle(val_data)
suffled_val_data
```

```python
splited_data = np.split(suffled_data, (2, 3), axis=1)
X = splited_data[0]
y = splited_data[1]
print(X)
print(y)
```

```python
splited_val_data = np.split(suffled_val_data, (2, 3), axis=1)
X_val = splited_val_data[0]
y_val = splited_val_data[1]
print(X_val)
print(y_val)
```

```python
from layer import BasicNeuralNetwork

in_dim_n = 2 
hidden_dim_n = 8 
out_dim_n = 1
epoch = 10
learning_rate = 0.01
batch_size = 1
clip_grad_l2 = False
optim_fn = 'sgd'

bnn_sgd = BasicNeuralNetwork(
    in_dim_n, hidden_dim_n, out_dim_n, epoch, learning_rate, batch_size, optim_fn, clip_grad_l2)
```

```python
result, val_result = bnn_sgd.train(X, y, X_val, y_val)
plt.figure(figsize=(10, 8))
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(result, label='Train set')
plt.plot(val_result, label='Valid set')
plt.show()
```

```python
optim_fn = 'adagrad'

bnn_adagrad = BasicNeuralNetwork(
    in_dim_n, hidden_dim_n, out_dim_n, epoch, learning_rate, batch_size, optim_fn, clip_grad_l2)
result, val_result = bnn_adagrad.train(X, y, X_val, y_val)

plt.figure(figsize=(10, 8))
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(result, label='Train set')
plt.plot(val_result, label='Valid set')
plt.show()
```
