```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = np.loadtxt('training.txt')
data
```

```python
def suffle(x):
    arr = np.arange(0, len(x))
    np.random.shuffle(arr)
    return x[arr]   
```

```python
suffled_data = suffle(data)
suffled_data
```

```python
splited_data = np.split(suffled_data, (2, 3), axis=1)
X = splited_data[0]
y = splited_data[1]
X
```
